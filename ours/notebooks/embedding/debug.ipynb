```python
import pickle
from datasets import Dataset

corpus_path = '/home/tommy/Project/PCBSDA/ours/outputs/raw_data/embedding/pcode_corpus_x86_64_new_data.pkl'


def gen():
    with open(corpus_path, 'rb') as f:
        corpus = pickle.load(f)
        for sentence_tokens in corpus:
            yield {'sentence_tokens': sentence_tokens}
            # print(sentence_tokens)


dataset = Dataset.from_generator(gen)
print(dataset[1])    

```
