```python
import pickle
import networkx as nx

fpath = "/home/tommy/Project/PCBSDA/experiment/outputs/raw_data/single-architecture/FCGAT/GNN/gpickle_single/00/00ca815b1337d81a4b83a68abfd6237b1fe0dccd22cfae17e8a75c721f7fe9d5.gpickle"

with open(fpath, "rb") as f:
    G = pickle.load(f)

print(f"Type: {type(G)}")
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
print(f"Is directed: {G.is_directed()}")

# 查看前幾個 node 的屬性
print("\n--- Node attributes (first 3 nodes) ---")
for i, (node, attrs) in enumerate(G.nodes(data=True)):
    print(f"Node: {node}")
    print(f"  Attrs keys: {list(attrs.keys())}")
    for k, v in attrs.items():
        val_str = str(v)[:200]
        print(f"  {k}: {val_str}")
    if i >= 2:
        break

# 查看 graph-level 屬性
print(f"\n--- Graph attributes ---")
print(G.graph)

# 查看前幾條 edge
print("\n--- Edges (first 3) ---")
for i, (u, v, attrs) in enumerate(G.edges(data=True)):
    print(f"  {u} -> {v}  attrs: {attrs}")
    if i >= 2:
        break
```

```python
import pickle
import numpy as np

fpath_emb = "/home/tommy/Project/PCBSDA/experiment/outputs/embedded_graphs/single-architecture/FCGAT/00/00ca815b1337d81a4b83a68abfd6237b1fe0dccd22cfae17e8a75c721f7fe9d5.gpickle"

with open(fpath_emb, "rb") as f:
    G_emb = pickle.load(f)

print(f"Type: {type(G_emb)}")
print(f"Nodes: {G_emb.number_of_nodes()}")
print(f"Edges: {G_emb.number_of_edges()}")
print(f"Is directed: {G_emb.is_directed()}")

print("\n--- Graph attributes ---")
print(G_emb.graph)

print("\n--- Node attributes (first 3 nodes) ---")
for i, (node, attrs) in enumerate(G_emb.nodes(data=True)):
    print(f"Node: {node}")
    print(f"  Attrs keys: {list(attrs.keys())}")
    for k, v in attrs.items():
        if isinstance(v, np.ndarray):
            print(f"  {k}: ndarray shape={v.shape}, dtype={v.dtype}, values[:5]={v.flat[:5]}")
        else:
            print(f"  {k}: {str(v)[:200]}")
    if i >= 2:
        break

print("\n--- Edges (first 3) ---")
for i, (u, v, attrs) in enumerate(G_emb.edges(data=True)):
    print(f"  {u} -> {v}  attrs: {attrs}")
    if i >= 2:
        break
```
